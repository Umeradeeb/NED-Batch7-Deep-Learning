```python
import cv2
from matplotlib import pyplot as plt
# Read the image
img = cv2.imread('/content/cat.jpeg')

# Convert BGR (opencv default) to RGB for proper display with matplot
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.imshow(img_rgb)
plt.axis('off')   # Hide axis
plt.title('Output')
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blurr = cv2.GaussianBlur(img, (55, 55), 0)
print('Original Image Shape', img.shape)
print('Grayscale Image Shape', img_grey.shape)
print('Blur Image Shape', img_blurr.shape)

plt.figure(figsize=(20,10))

plt.subplot(2,3,1)
plt.imshow(img)
plt.title('Original Image BGR')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(img_rgb)
plt.title('RGB Image')
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(img_grey, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(img_blurr)
plt.title('Blurr Image')
plt.axis('off')


plt.show
```

```python
low_threshold = 50
high_threshold = 100
img_canny = cv2.Canny(img, low_threshold, high_threshold)

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image BGR')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_canny)
plt.title('Edge Detect Image')
plt.axis('off')

plt.show()
```

```python
img_resize = cv2.resize(img, (500, 500))


print('Original Image Shape', img.shape)
print('Resize Image Shape', img_resize.shape)

plt.figure(figsize=(15,6))

# plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image BGR')
plt.axis('off')
plt.show()

# plt.subplot(1,2,2)
plt.figure(figsize=(15,6))
plt.imshow(img_resize)
plt.title('Resize Image')
plt.axis('off')

plt.show()
```

```python
img_crop = img[0:2000,2000:3500]


print('Original Image Shape', img.shape)
print('Crop Image Shape', img_crop.shape)

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image BGR')
plt.axis('off')
# plt.show()

plt.subplot(1,2,2)
# plt.figure(figsize=(15,6))
plt.imshow(img_crop)
plt.title('Crop Image')
plt.axis('off')

plt.show()
```
